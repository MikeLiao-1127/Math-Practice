```python
#1.列出1加到100的總和過程
for i in range(1,101):
    total = 0
    for j in range(1, i+1):
        total = total + j
    print(total)
    
```

```python
#2.列出1乘到10的階乘表

for i in range(1, 11):
    total = 1
    for j in range(1,i+1):
        total = total * j
    print(i,'!=',total)


```

```python
#4.一年級新生有200名，分別編予711001、711002……711200的學號，現在請用程式列出學號一覽表
import pandas as pd
import numpy as np
number = np.arange(711001,711201)
student = np.arange(1,201)
#列舉學號
df1 = pd.DataFrame(number,columns=['學號'])
#列舉座號
df2 = pd.DataFrame(student,columns=['座號'])
#合併學號與座號
df3 = pd.concat([df1,df2], axis = 1)
df3
```

```python
#6.在for迴圈裡加入if條件判斷, 找500以內所有不能被3整除的數,將之加總.
total = 0
for i in range(1,501):
    if i %3 != 0:
        total = total + i 
print(total)
```

```python
#8.請建構一個1x10每一格都為0的矩陣A. 矩陣A每一格的編號為X,(X=1,2,…,10).
#每一格的數值以公式Y=X2+2X+3所計算出的新值Y來取代.請以for 迴圈完成此題.
import numpy as np
x = np.zeros((1,10), dtype = int)
for y in x:
    y = x*x+2*x+3
print(y)
```

```python
#9.請建構一個10x1每一格都為1的矩陣B. 矩陣B每一格的編號為X,(X=1,2,…,10). 
#每一格的數值以公式Y=X2+3X+5的新值Y來取代.請以for 迴圈完成此題.
import numpy as np
x = np.ones((10,1), dtype = int)
for y in x :
    y = x*x+3*x+5
print(y)
```
